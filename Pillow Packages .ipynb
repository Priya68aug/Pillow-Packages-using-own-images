```python
# To Flip the images using Pil 
import PIL
from PIL import Image
Out =cat_image.transpose(PIL.Image.FLIP_TOP_BOTTOM)
Out.save('transpose-output.jpg')
```

```python
#resizing the image
cat_image = Image.open("D:/image/cat.jpg")
#image size
size=(200,200)
#resize image
resized = cat_image.resize(size)
#save resized image
resized.save('resize-output.png')
```

```python
# now we will use the attribute box and give some values to it 
cat_image = Image.open("D:/image/cat.jpg")

#image size
size = (200,200)
box = (100,100,300,300)
#resize image
output_box = cat_image.resize(size, box=box)
#save resized image
output_box.save('resize-output.png')
```
